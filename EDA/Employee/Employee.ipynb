```python
import numpy as np
import pandas as pd
df=pd.read_csv("Employee.csv")
print(df)
df.shape # Display the shape of the DataFrame
df.describe() # Display statistical summary
df.head() # Display the first few rows
```

```python
df.info() # Display summary information
```

```python
print(df["salary"],df["department"])
```

```python
#people withdrwn more than 50000
print(df[df["salary"]>50000])
```

```python
#people with role as frontend developer
print(df[df["designation"]=="Frontend Developer"])
```

```python
#people who get salary more than 50000 and are in IT department
print(df[(df["salary"]>50000) & (df["department"]=="IT")])
```

```python
#display count salary more than 50000
print(len(df[df["salary"]>50000]))
```

```python
#minimum salary and who is withdrawn it
min_salary = df["salary"].min()
print(df[df["salary"] == min_salary])
#maximum salary and who is withdrawn it
print(df[df["salary"] == df["salary"].max()])
```

```python
#print people sort salary in descending order 
print(df.sort_values(by="salary", ascending=False))
```

```python
# print people of each  city
print(df['city'].value_counts())
```

```python
#sum of salary column
print(df['salary'].sum())
```

```python
#row where salary is not available
print(df[df["salary"].isnull()])
```

```python
#fill values where salary is not available with average salary
avg_salary = df['salary'].mean()
print(avg_salary)
df.fillna(avg_salary, inplace=True)
print(df)
```

```python
#push df that updated data to csv file
df.to_csv("Employee.csv", index=False)
```

```python
#round to salary column to 2 decimal places
df['salary'] = df['salary'].round(2)
print(df)
```

```python
#add new column same bonus 5000 to dataset 
df['bonus'] = 5000
print(df)
```

```python
#drop bonus column
df.drop('bonus', axis=1, inplace=True)
print(df)
```

```python
#hike hr 10% to it department employees and 5% to it and add column as new salary
df['new_salary'] = df['salary']
df.loc[df['department'] == 'IT', 'salary'] *= 1.10
df.loc[df['department'] == 'HR', 'salary'] *= 1.05
print(df)
```

```python
df["department"] = df["department"].replace("IT", 1)
df["department"] = df["department"].replace("HR", 2)
df["department"] = df["department"].replace("Finance", 3)
df["department"] = df["department"].replace("Operations", 4)
df["department"] = df["department"].replace("Sales", 5)
print(df)
```

```python
df.to_csv("Employee.csv", index=False)
print(df)
```

```python
#join 2 csv files Employee.csv and Department.csv df based on department column 
df_department = pd.read_csv("Department.csv")
merged_df = pd.merge(df, df_department, on="department", how="inner")
print(merged_df)
```

```python
df3=pd.read_csv("Department.csv")
df4=pd.merge(df, df3, left_on="department", right_on="department") #inner join
print(df4)
```
