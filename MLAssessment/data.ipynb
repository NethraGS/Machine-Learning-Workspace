```python
from io import StringIO

# Read and strip quotes from each line
with open('CustomerData.csv', 'r') as f:
    lines = [line.strip().strip('"') for line in f.readlines()]

csv_string = '\n'.join(lines)

df = pd.read_csv(StringIO(csv_string))
df
```

```python
# Fill numeric columns with mean
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean())

# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include=['object', 'string']).columns
for col in categorical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])

# Verify no nulls remain
print("Null values after filling:")
print(df.isnull().sum())
```

```python
print(f"Duplicate rows: {df.duplicated().sum()}")
```

```python
# Detect and remove outliers using IQR method for numerical columns (union of all outliers)
numerical_cols = ['AnnualIncome', 'TotalSpent', 'MonthlyPurchases', 'AvgOrderValue', 'AppTimeMinutes']

outlier_indices = set()

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {len(col_outliers)} outliers detected")
    outlier_indices.update(col_outliers.index)

outliers_df = df.loc[list(outlier_indices)]

# Save outliers to separate CSV
outliers_df.to_csv('outliers_dataset.csv', index=False)
print(f"\n{'='*50}")
print(f"Total unique outliers (union): {len(outlier_indices)}")
print(f"Outliers saved to 'outliers_dataset.csv'")

# Remove outliers from main dataset
df = df.drop(outlier_indices)
print(f"\nMain dataset after outlier removal:")
print(f"Shape: {df.shape}")
print(f"Remaining customers: {len(df)}")
print('='*50)
```

```python
# Display outliers summary
print("\nOutliers Summary:")
print(f"Total outliers removed: {len(outliers_df)}")
print("\nTop 5 outliers by TotalSpent:")
print(outliers_df.nlargest(5, 'TotalSpent')[['CustomerID', 'TotalSpent', 'AnnualIncome', 'MonthlyPurchases', 'AvgOrderValue']])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder

df_cluster = df.copy()

# Encode categorical variables
le_gender = LabelEncoder()
df_cluster['Gender_Encoded'] = le_gender.fit_transform(df_cluster['Gender'])

le_discount = LabelEncoder()
df_cluster['DiscountUsage_Encoded'] = le_discount.fit_transform(df_cluster['DiscountUsage'])

le_time = LabelEncoder()
df_cluster['PreferredShoppingTime_Encoded'] = le_time.fit_transform(df_cluster['PreferredShoppingTime'])


print("Encoding completed!")
print(f"Gender mapping: {dict(zip(le_gender.classes_, le_gender.transform(le_gender.classes_)))}")
print(f"DiscountUsage mapping: {dict(zip(le_discount.classes_, le_discount.transform(le_discount.classes_)))}")
print(f"PreferredShoppingTime mapping: {dict(zip(le_time.classes_, le_time.transform(le_time.classes_)))}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features = ['Age', 'Gender_Encoded', 'AnnualIncome', 'TotalSpent', 'AvgOrderValue', 
            'MonthlyPurchases', 'DiscountUsage_Encoded', 'AppTimeMinutes', 
            'PreferredShoppingTime_Encoded']

X = df_cluster[features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"Features selected for clustering: {features}")
print(f"Number of features: {len(features)}")
print(f"Scaled data shape: {X_scaled.shape}")
```

```python
# Apply K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_cluster['Cluster'] = kmeans.fit_predict(X_scaled)

df['Cluster'] = df_cluster['Cluster']

print(f"Clustering completed!")
print(f"\nCluster distribution:")
print(df['Cluster'].value_counts().sort_index())
```

```python
# CRITICAL: Remap clusters based on spending patterns
# Ensure Cluster 0 = High-Value, Cluster 1 = Regular, Cluster 2 = Price-Sensitive

# Analyze current clusters by average spending
cluster_spending = df.groupby('Cluster')['TotalSpent'].mean().sort_values(ascending=False)
print("Current cluster spending (sorted):")
print(cluster_spending)
print()

# Create mapping: highest spending → 0, medium → 1, lowest → 2
spending_ranks = cluster_spending.index.tolist()
cluster_mapping = {
    spending_ranks[0]: 0,  # Highest spending → Cluster 0 (High-Value Loyal)
    spending_ranks[1]: 1,  # Medium spending → Cluster 1 (Value-Seeking Regular)
    spending_ranks[2]: 2   # Lowest spending → Cluster 2 (Price-Sensitive Occasional)
}

print("Cluster Remapping:")
print(f"Old Cluster {spending_ranks[0]} (Highest Spending) → New Cluster 0 (High-Value Loyal)")
print(f"Old Cluster {spending_ranks[1]} (Medium Spending) → New Cluster 1 (Value-Seeking Regular)")
print(f"Old Cluster {spending_ranks[2]} (Lowest Spending) → New Cluster 2 (Price-Sensitive Occasional)")
print()

# Apply the remapping
df['Cluster'] = df['Cluster'].map(cluster_mapping)
df_cluster['Cluster'] = df_cluster['Cluster'].map(cluster_mapping)

print("✓ Clusters remapped successfully!")
print(f"\nNew cluster distribution:")
print(df['Cluster'].value_counts().sort_index())
```

```python
# Comprehensive cluster analysis with all features
cluster_analysis = df.groupby('Cluster').agg({
    'Age': 'mean',
    'AnnualIncome': 'mean',
    'TotalSpent': 'mean',
    'AvgOrderValue': 'mean',
    'MonthlyPurchases': 'mean',
    'AppTimeMinutes': 'mean',
    'CustomerID': 'count'
}).round(2)

cluster_analysis.columns = ['Avg_Age', 'Avg_AnnualIncome', 'Avg_TotalSpent', 'Avg_OrderValue', 
                             'Avg_MonthlyPurchases', 'Avg_AppTime', 'Customer_Count']

print("="*80)
print("COMPREHENSIVE CLUSTER ANALYSIS")
print("="*80)
print(cluster_analysis)
print("\n" + "="*80)
```

```python
# Analyze categorical variables by cluster
print("\n" + "="*40)
print("GENDER DISTRIBUTION BY CLUSTER (%)")
print("="*40)
gender_analysis = pd.crosstab(df['Cluster'], df['Gender'], normalize='index') * 100
print(gender_analysis.round(2))

print("\n" + "="*40)
print("DISCOUNT USAGE BY CLUSTER (%)")
print("="*40)
discount_analysis = pd.crosstab(df['Cluster'], df['DiscountUsage'], normalize='index') * 100
print(discount_analysis.round(2))

print("\n" + "="*50)
print("PREFERRED SHOPPING TIME BY CLUSTER (%)")
print("="*50)
time_analysis = pd.crosstab(df['Cluster'], df['PreferredShoppingTime'], normalize='index') * 100
print(time_analysis.round(2))
print("="*80)
```

```python
# Visualize clusters using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
sns.set_style("whitegrid")
sns.set_palette("husl")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Annual Income vs Total Spent
sns.scatterplot(data=df, x='AnnualIncome', y='TotalSpent', hue='Cluster', 
                palette='viridis', s=100, alpha=0.7, edgecolor='black', ax=axes[0, 0])
axes[0, 0].set_xlabel('Annual Income (₹)', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Total Spent (₹)', fontsize=12, fontweight='bold')
axes[0, 0].set_title('Clusters: Annual Income vs Total Spent', fontsize=14, fontweight='bold')
axes[0, 0].legend(title='Cluster', title_fontsize=11, fontsize=10)

# Plot 2: Monthly Purchases vs Avg Order Value
sns.scatterplot(data=df, x='MonthlyPurchases', y='AvgOrderValue', hue='Cluster',
                palette='viridis', s=100, alpha=0.7, edgecolor='black', ax=axes[0, 1])
axes[0, 1].set_xlabel('Monthly Purchases', fontsize=12, fontweight='bold')
axes[0, 1].set_ylabel('Average Order Value (₹)', fontsize=12, fontweight='bold')
axes[0, 1].set_title('Clusters: Monthly Purchases vs Avg Order Value', fontsize=14, fontweight='bold')
axes[0, 1].legend(title='Cluster', title_fontsize=11, fontsize=10)

# Plot 3: Average App Time by Cluster
cluster_app_time = df.groupby('Cluster')['AppTimeMinutes'].mean().reset_index()
sns.barplot(data=cluster_app_time, x='Cluster', y='AppTimeMinutes', 
            palette='coolwarm', edgecolor='black', ax=axes[1, 0])
axes[1, 0].set_xlabel('Cluster', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Avg App Time (minutes)', fontsize=12, fontweight='bold')
axes[1, 0].set_title('Average App Time by Cluster', fontsize=14, fontweight='bold')

# Add value labels on bars
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, fmt='%.1f', fontweight='bold')

# Plot 4: Average Total Spent by Cluster
cluster_spent = df.groupby('Cluster')['TotalSpent'].mean().reset_index()
sns.barplot(data=cluster_spent, x='Cluster', y='TotalSpent',
            palette='rocket', edgecolor='black', ax=axes[1, 1])
axes[1, 1].set_xlabel('Cluster', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('Avg Total Spent (₹)', fontsize=12, fontweight='bold')
axes[1, 1].set_title('Average Total Spent by Cluster', fontsize=14, fontweight='bold')

# Add value labels on bars
for container in axes[1, 1].containers:
    axes[1, 1].bar_label(container, fmt='%.0f', fontweight='bold')

plt.tight_layout()
plt.show()
```
